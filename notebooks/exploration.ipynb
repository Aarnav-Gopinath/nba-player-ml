```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: make plots look nicer
sns.set(style="whitegrid")
```

```python
# Load the CSV created from your pipeline
data = pd.read_csv("data/player_game_logs.csv")

# Quick look at the data
data.head()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data['MIN'], bins=20, kde=True)
plt.title("Distribution of Minutes Played")
plt.xlabel("Minutes")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data['PTS'], bins=20, kde=True)
plt.title("Distribution of Points Scored")
plt.xlabel("Points")
plt.ylabel("Count")
plt.show()
```

```python
# Select only numeric columns
numeric_data = data.select_dtypes(include=['number'])

# Compute correlations
corr = numeric_data.corr()

# Plot
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Feature Correlations")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='MIN', y='PTS', data=data)
plt.title("Minutes vs Points")
plt.xlabel("Minutes")
plt.ylabel("Points")
plt.show()
```
